```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(2000, 2)
v = [-0.5, -0.5]
w = [2, 2]
X += v
X *= w

print(X)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
r1 = np.concatenate([X[:2000,] < 0, X[:2000,] > 0])
r2 = np.concatenate([X[:2000,] > 0, X[:2000,] < 0])

b1 = np.concatenate([X[:2000,] < 0, X[:2000,] < 0])
b2 = np.concatenate([X[:2000,] > 0, X[:2000,] > 0])

plt.scatter(r1, r2, c='red')
plt.scatter(b1, b2, c='blue')
plt.show()
```
